```python
import sys
sys.path.append('./faster_RCNN/')
from stamp_dataset_class import StampDataset

import os
import torch
import pandas as pd
from IPython.display import clear_output

from engine import evaluate
import utils
import transforms as T

import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def get_transform(train):
    transforms = []
    # converts the image, a PIL image, into a PyTorch Tensor
    transforms.append(T.ToTensor())
    if train:
        # during training, randomly flip the training images
        # and ground-truth for data augmentation
        transforms.append(T.RandomHorizontalFlip(0.5))
    return T.Compose(transforms)
```

```python
import utils

test_df = pd.read_pickle('test_v4_df.pkl')
test_dataset = StampDataset(data_df=test_df, transforms=get_transform(train=False))

def inference_collate(batch):
    data = torch.cat([item[0][None, :] for item in batch], dim=0)
    return data
    
data_loader_test = torch.utils.data.DataLoader(dataset=test_dataset,
                                               batch_size=8,
                                               shuffle=False,
                                               num_workers=4,
                                               collate_fn=utils.collate_fn,
                                               pin_memory=True if torch.cuda.is_available() else False)
```

```python
from torchvision import transforms
import numpy as np
from tqdm.notebook import tqdm

@torch.no_grad()
def inference(model, data_loader, device):
    
    model.eval()
    model.to(device)
    preprocess = get_transform(train=False)
    
    preds = []
    with torch.no_grad():
        for i, (X_i, Y_i) in enumerate(tqdm(test_dataset)):
            X_i_GPU = X_i.to(device)[None, :]
            preds += [{k: v.cpu().detach().numpy() for k,v in res.items()} for res in model(X_i_GPU)]
            del X_i_GPU

    return preds
        
```

```python
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
checkpoint_path = 'trained_models_v4/checkpoint_v4-07.pt'

model = torch.load(checkpoint_path, map_location=device)
model.to(device)
preds = inference(model, data_loader_test, device)
```

```python
out_dir = 'out'
os.makedirs(out_dir, exist_ok=True)

for i, pred in enumerate(preds):
    im = cv2.imread(test_dataset.card_paths[i])
    for (x1, y1, x2, y2), score in zip(pred['boxes'], pred['scores']):
        if score >.3:
            cv2.rectangle(im, (int(x1), int(y1)), (int(x2), int(y2)), (23,46,111), 2)
    cv2.imwrite(os.path.join(out_dir, str(i).zfill(3) + '_' + str(score) + '.jpg'), im)
    
# for i, row in test_df.iterrows():
#     im = cv2.imread(row['path'])
#     x1, y1, x2, y2 = row['points']
#     cv2.rectangle(im, (x1, y1), (x2, y2), (23,46,111), 2)
#     cv2.imwrite(os.path.join(out_dir, str(i).zfill(3) + '_' + str(row['score']) + '.jpg'), im)
```

```python
SCORE_THRE = .05
    # idx = np.random.randint(len(test_dataset))
for idx in range(len(test_dataset)):
    boxes = preds[idx]['boxes']
    true_boxes = test_dataset[idx][1]['boxes'].numpy()
    scores = preds[idx]['scores']
    img_path  = test_dataset.card_paths[idx]
    names = test_dataset.data_df.query('path==@img_path')['name'].unique()
    response = None
    while response is None or response not in [0, 1]:
        im = cv2.imread(img_path)
        for box_id, ((x1, y1, x2, y2), score) in enumerate(zip(boxes, scores)):
            if score > SCORE_THRE:
                cv2.rectangle(im, (x1, y1), (x2, y2), (23,46,111), 10)
                cv2.rectangle(im, (x1, y1), (int(x1+100), int(y1+70)), (50,50,50), cv2.FILLED)
                cv2.putText(im, str(box_id+2), (int(x1), int(y1)+50), cv2.FONT_HERSHEY_PLAIN, 4, (200,200,0), 5)
        for (x1, y1, x2, y2) in true_boxes:
            p1, p2 = (int(x1+(x2-x1)/3), int(y1+(y2-y1)/3)), (int(x2-(x2-x1)/3), int(y2-(y2-y1)/3))
            cv2.rectangle(im, p1, p2, (111,46,23), 10)
        plt.figure(figsize=(20,10))
        plt.title('{}, {:.2f}'.format(idx, idx/len(test_dataset)))
        plt.imshow(im)
        plt.axis('off')
        plt.show()
        response = int(input('replace labels?'))
        if response==1:
            for name in names:
                dataset[name]['active_boxes_v4'] = []
                for (x1, y1, x2, y2), score in zip(boxes, scores):
                    if score > SCORE_THRE:
                        dataset[name]['active_boxes_v4'].append({'box_id': 0, 'box': [x1, y1, x2, y2]})
        if response == 0:
            for name in names:
                dataset[name]['active_boxes_v4'] = (dataset[name]['active_boxes_v3'] if 'active_boxes_v3' in dataset[name]
                                                    else dataset[name]['active_boxes_v2'])
        if response > 1:
            boxes = np.concatenate([boxes[:response-2], boxes[response-2+1:]])
            scores = np.concatenate([scores[:response-2], scores[response-2+1:]])
        clear_output(wait=True)
```

```python
for k, v in dataset.items():
    if 'active_boxes_v3' not in v:
        dataset[k]['active_boxes_v3'] = v['active_boxes_v2'] if 'active_boxes_v2' in v else v['active_boxes'] if 'active_boxes' in v else v['retrieved_boxes']
```
