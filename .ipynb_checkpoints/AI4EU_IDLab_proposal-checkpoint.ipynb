# Challenge Details

### DIEA: Development of Information Extraction Algorithms from Photographs

Challenge Owner: TOELT

Website URL: http://toelt.ai/

TOELT is a research company based in Switzerland that focuses on research on Artificial Intelligence applied in different sciences. It works bringing together infrastructure and technology partners (as NVIDIA) and research entities to work together on impactful research projects, bringing to the game research and artificial intelligence advanced expertise.

**Context**

Many companies struggle to optimise their processes and contracts because lots of information about clients, partners and processes are often stored in documents (PDFs, images, scans, etc.). Extracting text is not always easy, especially if the structure of said text plays a role. OCR methods do not always work, as scan quality, place on paper and many more factors influence the results that are normally not good enough for a wide-spread use of such methods. Sometimes information is not only stored in text, but in small images and in their position.

TOELT is collaborating in the construction and exploration of a new historical dataset on the Italian interwar labour market in partnership with Dr. Aurora Iannello, researcher in economic and social history at the Fondazione Luigi Einaudi in Turin. The purpose of the study is to perform an analysis of the social security cards stored at the archives of the National Institute for Social Security (INPS) and the Einaudi Foundation.

**Challenge**

The cards record the working history of Italian employees since 1920, when the subscription to the national system of social insurance became mandatory. A sample of several hundred private-sector employees active in the 1920s and 1930s will be selected and digitized from the INPS archives of Vercelli, Turin, Rome and Naples. The goal of this challenge is, by using this dataset, the development of new AI Based algorithms for the extraction of information, in this case, from images.

**Expected demonstration:**

This project has the main aim of developing new algorithms that will be able to be used to extract information from PDFs and images. The goal is not to build a better OCR (Optical Character Recognition) software, but build an AI based software that will understand the context of the texts and images and translate that into a digital format. Multiple objectives will be considered in increasing order of difficulty. For example, extract number and position of stamps based on the colour and pattern, extract specific information about persons (as birth date, etc.), train models to recognize handwritten text from the beginning of last century, and so on.

**Available datasets:**

A dataset of 5000 images of several hundred workers will be available. The privacy concerns will be addressed together with possible groups that will be selected for the challenge and with INPS (which owns the cards). The ultimate goal is to make all the data available and free to use for research purposes to everyone.

# Solution Proposal

## STAMP: Spatio-Temporal Analysis of Metadata in Pictures

#### The included powerpoint contains additional details about the proposal

 Within this project, we will focus on the detection and recognition of metadata objects that accompany social security cards. As these cards contain structured data, we aim to exploit this fact to extract valuable metadata.

First, the visual objects of interest (OOIs) shall be detected using generic state-of-the-art object detection or segmentation models. These objects include stamps, photographs, and other recognizable symbols. A small portion of the dataset will therefore be manually labeled with these objects. This is necessary to perform transfer learning of the detector models and also validate our results. For each type of OOI, we will also define some quality metrics and develop tools to automatically detect quality issues. Additionally, similar OOIs can be clustered based on visual similarity to perform near-duplicate detection and re-identification.
This will ease linking objects across multiple collections (if needed). 

In the next step, we aim to extract all printed text and lines from the documents and use this as a template form, which will later be enhanced with the handwritten text.  Once the printed text has been analyzed, the handwritten text (if any) can be recognized. For this step, we shall probably use a separate model, focused on handwritten text recognition (HTR). If we already know the area in which the handwritten text can appear, we can make a tight crop to improve the recognition performance. Many HTR models and datasets already exist, we shall investigate if these models can be reused without retraining. If they can't, we shall manually label part of the dataset ourselves to retrain them.

In the final step, all the extracted information will be combined and validated, before storing them in a central database. Fuzzy matching techniques will be explored to match the predicted texts to existing vocabularies and perform cross-collection linking. Each of these steps will be trained, evaluated, and optimized on the provided dataset.

### Solution pipeline

1. Label part of the dataset for model training & validation
2. Preprocess the scans (e.g. remove background, split into pages)
3. Detect and classify objects of interest on each card
4. Detect potential quality issues with the card and objects of interest
5. Use line detection and printed text to determine a template for the cards containing forms and tables
6. Detect handwritten text and fill in the template
7. Use fuzzy matching to match the predicted texts to existing vocabularies/databases and perform cross-collection linking
