```python
import os
import torch
from torch import nn
import pandas as pd
import numpy as np
import cv2
import albumentations as A
import torchvision.transforms as T
import matplotlib.pyplot as plt

DEVICE = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

```python
class HandDigitModel(nn.Module):
    def __init__(self):
        super(HandDigitModel, self).__init__()

        self.feature_extractor = torch.hub.load('pytorch/vision:v0.10.0', 'squeezenet1_0', pretrained=False)
        
        self.linear = nn.Sequential(nn.Linear(1000, 256), nn.ReLU(), nn.Linear(256, 11*6))

    def forward(self, x):
        x = self.feature_extractor(x)
        out = self.linear(x)
        out = out.view(len(x), 6, 11)
        return out
```

```python
test_transforms = A.Compose([
    A.Resize(128, 128)])
```

```python
model = HandDigitModel()
model.load_state_dict(torch.load('./text_recognition/HandDigitNet-checkpoint-v1.pt', map_location=DEVICE))
model.to(DEVICE)
print('Model loaded successfully!')
```

```python
IMAGE_FOLDER = '../download_from_drive/data/ProcessedO7/'
test_df = pd.read_pickle('./text_test_df.pkl')

for _ in range(40):
    row = test_df.sample(1).iloc[0]
    img = cv2.cvtColor(cv2.imread(os.path.join(IMAGE_FOLDER, row['filename'])), cv2.COLOR_RGB2GRAY)
    x1, y1, x2, y2 = row['points']
    print('Actual    Label:', row['value'])
    img = img[y1:y2, x1:x2]
    X_i = test_transforms(image=img)['image']
    X_i = T.ToTensor()(X_i)
    X_i = torch.cat([X_i]*3, dim=0)[None]
    model.eval()
    with torch.no_grad():
        X_i_gpu = X_i.to(DEVICE)
        preds = model(X_i_gpu).detach().cpu().numpy()
        del X_i_gpu
    pred_digits = np.argmax(preds[0], axis=1)
    string_result = ''.join([str(el) for i, el in enumerate(pred_digits) if 10 not in pred_digits[:i+1]])
    if len(string_result)==6:
        string_result = string_result[:2]+'-'+string_result[2:4]+'-'+string_result[4:]
    print('Predicted Label:', string_result if len(string_result)>0 else 'none')
    plt.imshow(img, cmap='gray'); plt.axis("off"); plt.show()
    print("\n\n")
```
