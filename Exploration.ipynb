```python
import urllib.request

def error_handler(f):
    def inner_func(*args, **kwargs):
        try:
            res = f(*args, **kwargs)
            return res, True
        except Exception as e:
            return None, False
    return inner_func

@error_handler
def download_video(video_url, video_path):
    urllib.request.urlretrieve(video_url, video_path)
```
