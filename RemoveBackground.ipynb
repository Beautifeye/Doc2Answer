```python
def extract_background_from_card(img_path, resize_factor = 10):
    fn = img_path.split('/')[-1]
    dst_path = os.path.join(dst_folder, fn)
    if os.path.isfile(dst_path):
        return
#     print(img_path)
    original_img = cv2.imread(img_path)
    height, width = original_img.shape[:2]
    img = cv2.resize(original_img, (width//resize_factor, height//resize_factor))
    cv2.rectangle(img, (0,0), (img.shape[1], img.shape[0]), (0,0,0), 50)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#     fig, axes = plt.subplots(1, 3, figsize=(25,10))
#     ax1, ax2, ax3 = axes
#     plt.suptitle(str(n)+' | '+ img_path,y=.75)
#     ax1.imshow(img)

    mask1 = (img[:,:,1] > img[:,:,0]*(1.02-0.24)) * (img[:,:,1] < img[:,:,0]*(1.02+0.24))
    mask2 = (img[:,:,2] > img[:,:,0]*(1.265 - 0.28)) * (img[:,:,1] < img[:,:,0]*(1.2659+0.28))
    mask3 = (img[:,:,0] > 160) * (img[:,:,1] > 160) * (img[:,:,2] > 160)
    mask = (1 - mask1 * mask2) * mask3
#     ax2.imshow(mask)

    # x_mean = int((np.arange(602) * mask.max(axis=0)).sum() / mask.max(axis=0).sum())
    # y_mean = int((np.arange(402) * mask.max(axis=1)).sum() / mask.max(axis=1).sum())
    # h, w = 90, 130
    # ax2.imshow(img[y_mean-h:y_mean+h, x_mean-w:x_mean+w])

    x_ticks = [x_tick if value else -1 for x_tick, value in zip(np.arange(602), mask.max(axis=0))]
    y_ticks = [y_tick if value else -1 for y_tick, value in zip(np.arange(402), mask.max(axis=1))]
    x_ticks = [tick for i, tick in enumerate(x_ticks) if np.median(x_ticks[max(0,i-10):min(i+10, len(x_ticks))])>0 and tick>=0]
    y_ticks = [tick for i, tick in enumerate(y_ticks) if np.median(y_ticks[max(0,i-10):min(i+10, len(y_ticks))])>0 and tick>=0]

    x1 = int(np.percentile(x_ticks, 1)) - 10
    x2 = int(np.percentile(x_ticks, 99)) + 10

    y1 = int(np.percentile(y_ticks, 1)) - 10
    y2 = int(np.percentile(y_ticks, 99)) + 10

#     ax3.imshow(img[y1:y2, x1:x2])
    cv2.imwrite(dst_path, original_img[y1*10:y2*10, x1*10:x2*10])
#     plt.show()
```

```python
img = cv2.imread(card_paths[120])
height, width = img.shape[:2]
img = cv2.resize(img, (width//10, height//10))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img); plt.show()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (31, 31), .8)
edge = cv2.Canny(blur, 0, 50, 3)
plt.imshow(blur); plt.show()
```

```python
contours, hierarchy = cv2.findContours(edge, cv2.RETR_EXTERNAL,
                                       cv2.CHAIN_APPROX_SIMPLE)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX    

for contour, hier in zip(contours, hierarchy):
    (x,y,w,h) = cv2.boundingRect(contour)
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 20)
    cv2.putText(img, ('width = {}, height = {}'.format(w, h)),
                (x+3, y+3),
                font,
                .1,
                (0, 255, 0),
                1,
                cv2.LINE_AA)

plt.imshow(img)
```
